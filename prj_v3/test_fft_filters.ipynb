```python
import numpy as np
import skimage

from viewer import Viewer
from image import *
from fft import *
```

```python
viewer = Viewer().set_unit(2.5)
raw = Image(skimage.data.astronaut()).set_title("Original")
img = Gray(raw).info()

fft = FFT2DShift(img)
amp = Amplitude(fft).info()
ang = Phase(fft).info()
inv = Abs(InvFFT2DShift(fft)).info()

viewer.show(raw, Log1p(amp), ang, inv)
```

```python
fft = FFT2D(img)
amp = Amplitude(fft).info()
ang = Phase(fft).info()
inv = Abs(InvFFT2D(fft)).info()

viewer.show(raw, Log1p(amp), ang, inv)
```

```python
img1 = Gray(Image(skimage.data.astronaut())).data_info().set_title("Astronaut").set_axis_off(True)
img2 = Rescale(Image(skimage.data.camera()), (0, 1)).data_info().set_title("Camera Man").set_axis_off(True)

fft1 = FFT2DShift(img1)
amp1 = Amplitude(fft1)
ang1 = Phase(fft1)

fft2 = FFT2DShift(img2)
amp2 = Amplitude(fft2)
ang2 = Phase(fft2)

viewer.show(img1, Log1p(amp1), ang1)
viewer.show(img2, Log1p(amp2), ang2)
```

```python
img1_inv = Abs(InvFFT2DShift(amp1, ang2)).info().set_axis_off(True)
img2_inv = Abs(InvFFT2DShift(amp2, ang1)).info().set_axis_off(True)

viewer.reset().set_unit(2.5)
viewer.show(img1, Log1p(amp1), ang2, img1_inv)
viewer.show(img2, Log1p(amp2), ang1, img2_inv)
```
